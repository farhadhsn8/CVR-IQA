```python
with open('stacking_mode_losses/train_acc.txt', 'r') as file:
    train_acc = [float(num) for num in file.read().split(' , ') if num]

with open('stacking_mode_losses/test_acc.txt', 'r') as file:
    test_acc = [float(num) for num in file.read().split(' , ') if num]



import matplotlib.pyplot as plt

# Plotting training accuracy
plt.figure(figsize=(10,5))
plt.plot(train_acc)
plt.title('Train SROCC')
plt.xlabel('Epochs')
plt.ylabel('SROCC')
plt.show()

# Plotting testing accuracy
plt.figure(figsize=(10,5))
plt.plot(test_acc)
plt.title('Test SROCC')
plt.xlabel('Epochs')
plt.ylabel('SROCC')
plt.show()



```
